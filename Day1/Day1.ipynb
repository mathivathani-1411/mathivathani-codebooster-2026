```python
cs_students=df[df['department']=='Computer Science']
high_attendance=df[df['attendance_percentage']>90]
print(high_attendance[['name','attendance_percentage']])
```

```python
df['total_score']=df['math_score']+df['programming_score']+df['science_score']+ df['english_score']
average= df[df['total_score']/4 > 90]
print(average[['name','total_score']])
```

```python
average_depart= df.groupby('department')['total_score'].mean().round(1)
print("\nAverage Per Department:")
print(average_depart)
```

```python
average_depart= df.groupby('department')['total_score'].mean().round(1).sort_values(ascending=False)
print("\nAverage Per Department:")
print(average_depart)
```
