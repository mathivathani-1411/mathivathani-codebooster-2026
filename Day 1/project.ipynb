```python
import pandas as pd
```

```python
df=pd.read_csv('student_performance.csv')
```

```python
print("Total Students: ",df.shape[0])
```

```python
print("Total Columns: ", df.shape[1])
```

```python
print("Departments: ")
print(df['department'].nunique())
```

```python
print("No.Of.Students: ")
print(df['department'].value_counts())
```

```python
print("High Math score: ")
print(df['math_score'].max())
```

```python
print("low Math score: ")
print(df['math_score'].min())
```

```python
print("Average Math score: ")
print(df['math_score'].mean().round(2))
```

```python
print("Average Programming  score: ")
print(df['programming_score'].mean().round(2))
```

```python
print("Average Attendance: ")
print(df['attendance_percentage'].mean().round(2))
```

```python
df['total_score']=df['math_score']+df['programming_score']+df['science_score']+ df['english_score']
```

```python
print("\nGender Wise Average Score: ")
gender_scores = df.groupby('gender')['total_score'].mean().round(2)
print(gender_scores)
```

```python
top5=df.sort_values('total_score',ascending=False).head(5)
for index, row in top5.iterrows():
  print(f" {row['name']:<20} | {row['department']:<18} | Score: {row['total_score']}")
```

```python
low_att=df[df['attendance_percentage']<75][['name','department','attendance_percentage'] ]
if len(low_att)>0:
  print(low_att.to_string(index=False))

else:
  print("All students have attendance above 75%")

  print("\n"+"="*60)
  print("Report Complete")
  print("="*60)
```
